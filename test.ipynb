```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import graphviz

# Load the Iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Train a decision tree classifier
clf = DecisionTreeClassifier()
clf = clf.fit(X, y)

# Visualize the decision tree
dot_data = tree.export_graphviz(clf, out_file=None,
                     feature_names=iris.feature_names,
                     class_names=iris.target_names,
                     filled=True, rounded=True,
                     special_characters=True)
graph = graphviz.Source(dot_data)
graph.render("iris_decision_tree")  # Save the visualization to a file
graph  # Display the decision tree in Jupyter Notebook or IPython
```

```python
import math

def gcd_float(*args):
    print(type(args))
    # Scale the numbers to convert them to integers
    scaled_args = [int(arg * 10**10) for arg in args]
    # Calculate the GCD of scaled numbers
    gcd_scaled = math.gcd(*scaled_args)
    # Adjust the GCD to the original scale
    gcd_float = gcd_scaled / 10**10
    return gcd_float

# Example usage
num1 = 0.5
num2 = 0.75
num3 = 1.25
result = gcd_float(num1, num2, num3)
print("GCD:", result)
```

```python
import math

def largest_divisor(numbers):
    """
    Find the largest number that a set of numbers is divisible by.

    Parameters:
        numbers (list or array-like): The set of numbers.

    Returns:
        int: The largest number that the set of numbers is divisible by.
    """
    if len(numbers) == 0:
        return None

    gcd = numbers[0]
    for num in numbers[1:]:
        gcd = math.gcd(gcd, num)

    return gcd

# Example usage:
numbers = [12, 18, 24, 36]
largest_div = largest_divisor(numbers)
print("The largest number that the set of numbers is divisible by is:", largest_div)
```

```python
from graphviz import Digraph

class CustomTree:
    def __init__(self, data):
        self.data = data
        self.children = []

    def add_child(self, child):
        self.children.append(child)

    def visualize(self):
        dot = Digraph()
        self._visualize_helper(dot)
        dot.render('custom_tree', format='png', cleanup=True)

    def _visualize_helper(self, dot, parent_label=None):
        if parent_label is None:
            parent_label = str(self.data)

        dot.node(parent_label, label=str(self.data))

        for i, child in enumerate(self.children):
            child_label = f"{parent_label}_{i}"
            child._visualize_helper(dot, parent_label=child_label)
            dot.edge(parent_label, child_label)

# Example usage:
# Create custom tree
root = CustomTree("Root")
child1 = CustomTree("Child 1")
child2 = CustomTree("Child 2")
grandchild1 = CustomTree("Grandchild 1")
grandchild2 = CustomTree("Grandchild 2")

root.add_child(child1)
root.add_child(child2)
child1.add_child(grandchild1)
child2.add_child(grandchild2)

# Visualize the tree
root.visualize()
```

```python
import random

# Define the outcomes and their probabilities
outcomes = [
    ("Outcome 1", 0.7),  # 70% probability
    ("Outcome 2", 0.2),  # 20% probability
    ("Outcome 3", 0.08), # 8% probability
    ("Outcome 4", 0.02)  # 2% probability
]

# Generate a random number between 0 and 1
random_number = random.random()

# Determine the outcome based on the probabilities
cumulative_probability = 0
for outcome, probability in outcomes:
    cumulative_probability += probability
    if random_number < cumulative_probability:
        print(outcome)
        break
```
